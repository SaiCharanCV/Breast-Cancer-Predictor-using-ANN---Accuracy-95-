```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from imblearn.over_sampling import SMOTE
import pickle
```

```python
df = pd.read_csv(r"E:\Jupyter Notebook\Datasets\Breast Cancer Classification\breast-cancer.csv")
df
```

```python
df.diagnosis = pd.get_dummies(df.diagnosis, drop_first= True)
df
```

```python
x = df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean']]
y = df['diagnosis']
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x
```

```python
optimizer = Adam(learning_rate=0.001)
```

```python
df.isna().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,stratify=y, random_state=42)
len(x_train), len(x_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=50)

model.evaluate(x_test, y_test)

y_pred = model.predict(x_test)
y_pred_round = np.round(y_pred)
print(y_pred)
print(classification_report(y_test, y_pred_round))
```
